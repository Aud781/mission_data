```python
df= pd.read_csv(r'crime-in-france.csv')
```

```python
# 1. Melt pour déplier les colonnes de dates
df_melted = df.melt(id_vars=['Index', 'Libellé index'], 
                    var_name='Date', 
                    value_name='Valeur')

# 2. Pivot_table pour gérer les doublons
df_pivoted = df_melted.pivot_table(index='Date', 
                                   columns='Libellé index', 
                                   values='Valeur', 
                                   aggfunc='sum')  # Vous pouvez changer 'sum' par 'mean', 'max', etc.

# 3. Optionnel : Réinitialiser les noms des colonnes
df_pivoted.columns.name = None
df_pivoted.reset_index(inplace=True)

# Afficher le dataframe transformé
display(df_pivoted).head()
```
